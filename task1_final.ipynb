```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torchvision.models import resnet152, resnet18, ResNet152_Weights, ResNet18_Weights
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import TSNE
import umap
import time
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
from collections import defaultdict


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")


transform_train = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

trainset = torchvision.datasets.CIFAR10(
    root='./data', train=True, download=True, transform=transform_train)
trainloader = torch.utils.data.DataLoader(
    trainset, batch_size=128, shuffle=True, num_workers=2)

testset = torchvision.datasets.CIFAR10(
    root='./data', train=False, download=True, transform=transform_test)
testloader = torch.utils.data.DataLoader(
    testset, batch_size=100, shuffle=False, num_workers=2)


classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')



model = resnet152(weights=ResNet152_Weights.IMAGENET1K_V1)

for param in model.parameters():
    param.requires_grad = False

num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 10)
model.fc.requires_grad = True

model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.fc.parameters(), lr=0.001, momentum=0.9)

def train_model(model, trainloader, testloader, criterion, optimizer, num_epochs=5):
    train_losses = []
    train_accs = []
    val_losses = []
    val_accs = []

    for epoch in range(num_epochs):
        print(f'Epoch {epoch+1}/{num_epochs}')
        print('-' * 10)

        model.train()
        running_loss = 0.0
        running_corrects = 0

        for inputs, labels in trainloader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            loss = criterion(outputs, labels)

            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(trainset)
        epoch_acc = running_corrects.double() / len(trainset)

        train_losses.append(epoch_loss)
        train_accs.append(epoch_acc.cpu().numpy())

        print(f'Train Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        model.eval()
        running_loss = 0.0
        running_corrects = 0

        with torch.no_grad():
            for inputs, labels in testloader:
                inputs, labels = inputs.to(device), labels.to(device)

                outputs = model(inputs)
                _, preds = torch.max(outputs, 1)
                loss = criterion(outputs, labels)

                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(testset)
        epoch_acc = running_corrects.double() / len(testset)

        val_losses.append(epoch_loss)
        val_accs.append(epoch_acc.cpu().numpy())

        print(f'Val Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
        print()

    return train_losses, train_accs, val_losses, val_accs

print("Training baseline model (frozen backbone, only head trained)...")
train_losses, train_accs, val_losses, val_accs = train_model(
    model, trainloader, testloader, criterion, optimizer, num_epochs=5)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.title('Loss over epochs')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_accs, label='Training Accuracy')
plt.plot(val_accs, label='Validation Accuracy')
plt.title('Accuracy over epochs')
plt.legend()
plt.tight_layout()
plt.savefig('baseline_results.png')
plt.show()


def disable_skip_connections(model, layer_name, block_indices):
    """
    Disable skip connections in specified blocks of a ResNet layer
    """
    layer = getattr(model, layer_name)

    for idx in block_indices:
        if idx < len(layer):
            original_forward = layer[idx].forward

            def make_new_forward(block):
                def new_forward(x):
                    identity = x

                    out = block.conv1(x)
                    out = block.bn1(out)
                    out = block.relu(out)

                    out = block.conv2(out)
                    out = block.bn2(out)
                    out = block.relu(out)

                    out = block.conv3(out)
                    out = block.bn3(out)

                   
                    out = block.relu(out)
                    return out
                return new_forward

            layer[idx].forward = make_new_forward(layer[idx])

model_no_skip = resnet152(weights=ResNet152_Weights.IMAGENET1K_V1)

for param in model_no_skip.parameters():
    param.requires_grad = False

num_ftrs = model_no_skip.fc.in_features
model_no_skip.fc = nn.Linear(num_ftrs, 10)
model_no_skip.fc.requires_grad = True

disable_skip_connections(model_no_skip, 'layer4', [1, 2])

model_no_skip = model_no_skip.to(device)

optimizer_no_skip = optim.SGD(model_no_skip.fc.parameters(), lr=0.001, momentum=0.9)

print("Training model with disabled skip connections...")
train_losses_ns, train_accs_ns, val_losses_ns, val_accs_ns = train_model(
    model_no_skip, trainloader, testloader, criterion, optimizer_no_skip, num_epochs=5)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(train_losses, label='With Skip Connections')
plt.plot(train_losses_ns, label='Without Skip Connections')
plt.title('Training Loss Comparison')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(val_accs, label='With Skip Connections')
plt.plot(val_accs_ns, label='Without Skip Connections')
plt.title('Validation Accuracy Comparison')
plt.legend()
plt.tight_layout()
plt.savefig('skip_connection_comparison.png')
plt.show()


class FeatureExtractor:
    def __init__(self, model, layers):
        self.model = model
        self.layers = layers
        self.features = {layer: [] for layer in layers}
        self.hooks = []

        for name, module in model.named_modules():
            if name in layers:
                self.hooks.append(module.register_forward_hook(
                    lambda m, i, o, name=name: self.features[name].append(o.detach().cpu())
                ))

    def remove_hooks(self):
        for hook in self.hooks:
            hook.remove()

def extract_features(model, dataloader, layer_names, max_samples=1000):
    model.eval()
    extractor = FeatureExtractor(model, layer_names)
    labels_list = []

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(dataloader):
            if i * inputs.size(0) >= max_samples:
                break

            inputs = inputs.to(device)
            _ = model(inputs)
            labels_list.append(labels)

    features = {}
    for layer in layer_names:
        layer_features = torch.cat(extractor.features[layer], dim=0)

        if layer_features.dim() == 4:
            layer_features = torch.nn.functional.adaptive_avg_pool2d(layer_features, (1, 1))

        features[layer] = layer_features.view(layer_features.size(0), -1).numpy()

    extractor.remove_hooks()
    labels = torch.cat(labels_list, dim=0).numpy()

    return features, labels

layer_names = ['layer1', 'layer2', 'layer3', 'layer4', 'fc']

print("Extracting features from different layers...")
features, labels = extract_features(model, testloader, layer_names, max_samples=1000)

def visualize_features_tsne(features, labels, layer_name, title):
    tsne = TSNE(n_components=2, random_state=42)
    features_2d = tsne.fit_transform(features)

    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(features_2d[:, 0], features_2d[:, 1], c=labels,
                         cmap='tab10', alpha=0.6)
    plt.legend(*scatter.legend_elements(), title="Classes")
    plt.title(f't-SNE visualization of {title}')
    plt.savefig(f'tsne_{layer_name}.png')
    plt.show()

    return features_2d

def visualize_features_umap(features, labels, layer_name, title):
    reducer = umap.UMAP(random_state=42)
    features_2d = reducer.fit_transform(features)

    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(features_2d[:, 0], features_2d[:, 1], c=labels,
                         cmap='tab10', alpha=0.6)
    plt.legend(*scatter.legend_elements(), title="Classes")
    plt.title(f'UMAP visualization of {title}')
    plt.savefig(f'umap_{layer_name}.png')
    plt.show()

    return features_2d

def analyze_separability(features, labels, layer_names):
    separability_scores = {}

    for layer in layer_names:
        unique_classes = np.unique(labels)
        intra_distances = []
        inter_distances = []

        for cls in unique_classes:
            cls_indices = np.where(labels == cls)[0]
            cls_features = features[layer][cls_indices]

            if len(cls_indices) > 1:
                centroid = np.mean(cls_features, axis=0)
                distances = np.linalg.norm(cls_features - centroid, axis=1)
                intra_distances.extend(distances)

            next_cls = (cls + 1) % len(unique_classes)
            next_cls_indices = np.where(labels == next_cls)[0]
            next_cls_features = features[layer][next_cls_indices]

            if len(next_cls_indices) > 0:
                centroid1 = np.mean(cls_features, axis=0)
                centroid2 = np.mean(next_cls_features, axis=0)
                inter_distances.append(np.linalg.norm(centroid1 - centroid2))

        if intra_distances and inter_distances:
            separability = np.mean(inter_distances) / np.mean(intra_distances)
            separability_scores[layer] = separability
            print(f"Layer {layer}: Separability score = {separability:.4f}")

    return separability_scores

print("Analyzing class separability across layers...")
separability_scores = analyze_separability(features, labels, layer_names)

print("Visualizing features with t-SNE and UMAP...")
for layer in layer_names:
    if layer != 'fc':  
        print(f"Visualizing {layer} features...")
        tsne_result = visualize_features_tsne(features[layer], labels, layer, f'{layer} features')
        umap_result = visualize_features_umap(features[layer], labels, layer, f'{layer} features')

print("\n===== TRANSFER LEARNING AND GENERALIZATION =====")

def setup_model_for_finetuning(model, tune_strategy, weights=None):
    """
    Setup model for different fine-tuning strategies
    """
    if weights:
        model = resnet152(weights=weights)
    else:
        model = resnet152(weights=None)  
    num_ftrs = model.fc.in_features
    model.fc = nn.Linear(num_ftrs, 10)

    if tune_strategy == "head_only":
        for name, param in model.named_parameters():
            if not name.startswith('fc'):
                param.requires_grad = False
    elif tune_strategy == "last_block":
        for name, param in model.named_parameters():
            if not (name.startswith('layer4') or name.startswith('fc')):
                param.requires_grad = False
    elif tune_strategy == "full_backbone":
        for param in model.parameters():
            param.requires_grad = True

    return model.to(device)

strategies = ["head_only", "last_block", "full_backbone"]
results = {}

for strategy in strategies:
    print(f"\nTraining with strategy: {strategy} (ImageNet pretrained)")

    model_ft = setup_model_for_finetuning(None, strategy, ResNet152_Weights.IMAGENET1K_V1)

    if strategy == "head_only":
        optimizer_ft = optim.SGD(model_ft.fc.parameters(), lr=0.001, momentum=0.9)
    elif strategy == "last_block":
        optimizer_ft = optim.SGD([
            {'params': model_ft.fc.parameters(), 'lr': 0.001},
            {'params': model_ft.layer4.parameters(), 'lr': 0.0001}
        ], momentum=0.9)
    else:  
        fc_params = [p for p in model_ft.fc.parameters()]
        backbone_params = [p for n, p in model_ft.named_parameters() if not n.startswith('fc')]

        optimizer_ft = optim.SGD([
            {'params': fc_params, 'lr': 0.001},
            {'params': backbone_params, 'lr': 0.0001}
        ], momentum=0.9)

    train_losses, train_accs, val_losses, val_accs = train_model(
        model_ft, trainloader, testloader, criterion, optimizer_ft, num_epochs=5)

    results[strategy] = {
        'train_losses': train_losses,
        'train_accs': train_accs,
        'val_losses': val_losses,
        'val_accs': val_accs,
        'final_val_acc': val_accs[-1]
    }

print("\nTraining with random initialization (full_backbone strategy)")
model_random = setup_model_for_finetuning(None, "full_backbone", weights=None)
optimizer_random = optim.SGD(model_random.parameters(), lr=0.001, momentum=0.9)

train_losses_rand, train_accs_rand, val_losses_rand, val_accs_rand = train_model(
    model_random, trainloader, testloader, criterion, optimizer_random, num_epochs=5)

results["random_init"] = {
    'train_losses': train_losses_rand,
    'train_accs': train_accs_rand,
    'val_losses': val_losses_rand,
    'val_accs': val_accs_rand,
    'final_val_acc': val_accs_rand[-1]
}

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
for strategy in strategies:
    plt.plot(results[strategy]['val_accs'], label=strategy)
plt.plot(results["random_init"]['val_accs'], label='random_init')
plt.title('Validation Accuracy by Strategy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 3, 2)
for strategy in strategies:
    plt.plot(results[strategy]['val_losses'], label=strategy)
plt.plot(results["random_init"]['val_losses'], label='random_init')
plt.title('Validation Loss by Strategy')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 3, 3)
strategies_with_random = strategies + ["random_init"]
acc_values = [results[s]['final_val_acc'] for s in strategies_with_random]
plt.bar(strategies_with_random, acc_values)
plt.title('Final Validation Accuracy')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.savefig('transfer_learning_comparison.png')
plt.show()


def compare_visualization_methods(features, labels, layer_name):
    start_time = time.time()
    tsne = TSNE(n_components=2, random_state=42)
    features_tsne = tsne.fit_transform(features)
    tsne_time = time.time() - start_time

    start_time = time.time()
    reducer = umap.UMAP(random_state=42)
    features_umap = reducer.fit_transform(features)
    umap_time = time.time() - start_time

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    scatter1 = ax1.scatter(features_tsne[:, 0], features_tsne[:, 1], c=labels,
                          cmap='tab10', alpha=0.6)
    ax1.set_title(f't-SNE (time: {tsne_time:.2f}s)')
    ax1.legend(*scatter1.legend_elements(), title="Classes")

    scatter2 = ax2.scatter(features_umap[:, 0], features_umap[:, 1], c=labels,
                          cmap='tab10', alpha=0.6)
    ax2.set_title(f'UMAP (time: {umap_time:.2f}s)')
    ax2.legend(*scatter2.legend_elements(), title="Classes")

    plt.suptitle(f'Feature visualization comparison for {layer_name}')
    plt.savefig(f'comparison_{layer_name}.png')
    plt.show()

    return tsne_time, umap_time

print("Comparing t-SNE vs UMAP visualization methods...")
tsne_times = []
umap_times = []

for layer in ['layer1', 'layer4']:
    tsne_time, umap_time = compare_visualization_methods(features[layer], labels, layer)
    tsne_times.append(tsne_time)
    umap_times.append(umap_time)

print(f"t-SNE average time: {np.mean(tsne_times):.2f}s")
print(f"UMAP average time: {np.mean(umap_times):.2f}s")

def analyze_confusions(model, testloader):
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in testloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    cm = confusion_matrix(all_labels, all_preds)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=classes, yticklabels=classes)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig('confusion_matrix.png')
    plt.show()

    confusions = []
    for i in range(len(classes)):
        for j in range(len(classes)):
            if i != j and cm[i, j] > 0:
                confusions.append((i, j, cm[i, j]))

    confusions.sort(key=lambda x: x[2], reverse=True)

    print("\nTop confused class pairs:")
    for i, j, count in confusions[:10]:
        print(f"{classes[i]} mistaken as {classes[j]}: {count} times")

    return confusions

print("Analyzing model confusions...")
confusions = analyze_confusions(model, testloader)

print("\nComparing ResNet-152 vs ResNet-18 feature quality...")

model_resnet18 = resnet18(weights=ResNet18_Weights.IMAGENET1K_V1)
for param in model_resnet18.parameters():
    param.requires_grad = False

num_ftrs = model_resnet18.fc.in_features
model_resnet18.fc = nn.Linear(num_ftrs, 10)
model_resnet18.fc.requires_grad = True
model_resnet18 = model_resnet18.to(device)

features_18, labels_18 = extract_features(model_resnet18, testloader, ['layer4', 'fc'], max_samples=1000)

separability_152 = analyze_separability(features, labels, ['layer4', 'fc'])
separability_18 = analyze_separability(features_18, labels_18, ['layer4', 'fc'])

print("\nSeparability comparison:")
for layer in ['layer4', 'fc']:
    score_152 = separability_152.get(layer, 0)
    score_18 = separability_18.get(layer, 0)
    print(f"{layer}: ResNet-152 = {score_152:.4f}, ResNet-18 = {score_18:.4f}")

```
